```python
import pandas as pd 
import seaborn as sns
```

```python
# importing data frame 
dtf= sns.load_dataset('penguins')
dtf
```

```python
# renaming the columns 
dtf.columns= ['species', 'island', 'b_len', 'b_dep', 'f_len', 'bm_g', 'sex']
print(dtf.head())
```

```python
# correlation matrix form
matrix1= dtf.corr() 
print(matrix1)
```

```python
# make the decimals round 
matrix1= dtf.corr().round(2)
print(matrix1)
```

```python
#visualtion correlation 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# building the plot 
matrix1= dtf.corr().round(2) 
sns.heatmap(matrix1, annot=True)
plt.show
```

```python
# changing the color 
# building the plot 
matrix1= dtf.corr().round(2) 
sns.heatmap(matrix1, annot=True, vmax=1, vmin=-1, center=0, cmap='vlag')
plt.show
```

```python
# showing half of the heatmap 
matrix1= dtf.corr().round(2) 
mask= np.triu(np.ones_like(matrix1, dtype=bool))
sns.heatmap(matrix1, annot=True, vmax=1, vmin=-1, center=0, cmap='vlag', mask=mask)
plt.show
```
